```python
import pandas as pd
```

```python
data_import = pd.read_csv("data_cleaning_challenge.csv")
data_import
```

```python
drop_extra = data_import.drop(columns = ["Unnamed: 9","Unnamed: 10"])
drop_extra
```

```python
drop_nans = drop_extra[drop_extra["Row Type"].notna()]
drop_nans.head(20)
```

```python
col_values = []
counter = 0

for i in drop_nans["Row Type"]:
    if "first name" in i:
        counter += 1
    col_values.append(counter)
    
```

```python
# iter_call is a new data frame
iter_cols = drop_nans
iter_cols["Iteration"] = col_values
iter_cols
```

```python
drop_extra_col_rows = iter_cols[iter_cols["Row Type"] != "Row Types"]
drop_extra_col_rows
```

```python
name_df = drop_extra_col_rows[drop_extra_col_rows["Row Type"].str.contains("first name")]
name_df.drop(columns=["Speed1","Speed2","Electricity","Effort","Weight","Torque"], inplace= True)
name_df.rename(columns= {"Row Type": "First Name", "Iter Number" : "Last Name", "Power1":"Date"}, inplace= True)
name_df["First Name"] = name_df["First Name"].str[12:]
name_df["Last Name"] = name_df["Last Name"].str[11:]
name_df["Date"] = name_df["Date"].str[6:]
name_df
```

```python
no_name_df =  drop_extra_col_rows[~drop_extra_col_rows["Row Type"].str.contains("first name")]
no_name_df
```

```python
pd.merge(left = name_df, right = no_name_df, how = "inner", on = "Iteration")
```
